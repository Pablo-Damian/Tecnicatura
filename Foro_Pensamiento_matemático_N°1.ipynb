**Problema 1: "El hotel de los conjuntos"**

1 - El hotel de los conjuntos - consigna.png

```python
# Se definen las habitaciones
habitacionZ = set('abcdefghijklmnopqrstuvwxyz')
habitacionV = set('aeiou')
# Para simplificar, solo se consideran los números naturales y pares del 0 al 9
habitacionN = set(map(str, range(0, 10)))
habitacionP = set(map(str, range(0, 10, 2)))
habitacionC = set(['Córdoba', 'Corrientes', 'Formosa', 'La Plata', 'La Rioja', 'Mendoza', 'Neuquén', 'Paraná', 'Posadas', 'Rawson', 'Resistencia', 'Río Gallegos',
                    'Salta', 'San Fernando del Valle de Catamarca', 'San Juan', 'San Luis', 'San Miguel de Tucumán', 'San Salvador de Jujuy', 'Santa Fe', 'Santa Rosa',
                    'Santiago del Estero', 'Ushuaia', 'Viedma'])

# Se pide ingresar el nombre del "turista"
turista = input("Ingresar el turista: ")

# Se verifica en qué habitaciones puede ser alojado el "turista"
habitacionesPosibles = []
if turista in habitacionZ:
    habitacionesPosibles.append('Habitación Z')
if turista in habitacionV:
    habitacionesPosibles.append('Habitación V')
if turista in habitacionN:
    habitacionesPosibles.append('Habitación N')
if turista in habitacionP:
    habitacionesPosibles.append('Habitación P')
if turista in habitacionC:
    habitacionesPosibles.append('Habitación C')

# Si el "turista" no puede ser alojado en ninguna habitación, se agrega la opción "En ninguna habitación"
if not habitacionesPosibles:
    habitacionesPosibles.append('En ninguna habitación')

# Se imprimen las habitaciones posibles
print(f'El turista "{turista}" puede ser alojado en: {habitacionesPosibles}')

# Para el ejemplo se eligió la Pregunta 1 - ¿En cuál o cuáles habitaciones podría ser hospedada la letra "a"?
```

**Problema 1: "El hotel de los conjuntos numéricos"**

![1 - El hotel de los conjuntos numéricos - consigna.png]( )

```python
def asignarHabitacion(grupo):
    # Verificamos a qué grupo se pertenece
    if grupo == 1: # Números pares positivos e impares negativos
        return ['Habitación Z', 'Habitación Q', 'Habitación R']
    elif grupo == 2: # Números primos 2 y 3, y el número "π"
        return ['Habitación R']
    elif grupo == 3: # Raíces cuadradas de 3, 5, 7 y 11
        return ['Habitación I', 'Habitación R']
    elif grupo == 4: # Números que resultan de dividir a 1 por un número par positivo no nulo
        return ['Habitación Q', 'Habitación R']
    else: # Si no se pertenece a ningún grupo, no se aloja en ninguna habitación
        return ['En ninguna habitación']

# Se solicita al usuario que ingrese el número del grupo
grupo = int(input("Ingrese el número de grupo (1, 2, 3, 4): "))
habitacionesPosibles = asignarHabitacion(grupo)

# Finalmente se muestran las habitaciones en las que puede alojarse el grupo
print(f"El grupo {grupo} podría ser hospedado en: {', '.join(habitacionesPosibles)}")

# Para el ejemplo se eligió la Pregunta 4 - ¿En cuál o cuáles habitaciones podría ser hospedado el grupo 4?
```

**Problema 2: "El hotel de los números reales"**

![2 - El hotel de los números reales - consigna.png]( )

```python
# Valor de los huéspedes
valoresHuespedes = {
    "Huésped 1": 1.4142,
    "Huésped 2": 1273 / 900,
    "Huésped 3": 1.41422,  # con período en la cifra 2 (1.41422222...)
    "Huésped 4": 2**0.5,  # Raíz cuadrada de 2
}

# Los huéspedes se ordenan de izquierda a derecha basados en su valor numérico
huespedesOrdenados = sorted(valoresHuespedes.items(), key=lambda x: x[1])

# Se muestra/imprime el orden correcto de los huéspedes
for huesped, valor in huespedesOrdenados:
    print(f"{huesped}: {valor}")

# Este Problema consta de un solo ejercicio - De izquierda a derecha, ¿en qué orden quedarán ubicados los huéspedes en las habitaciones del hotel?
```
